```python
import pickle, pprint
from pathlib import Path

PKL = Path.cwd() / 'mat_eq_eq_lam_he.pkl'
with open(PKL, 'rb') as f:
    r = pickle.load(f)

print(f'=== {PKL.name} ===')
print(f'top-level keys: {list(r.keys())}\n')

for k, v in r.items():
    t = type(v).__name__
    if hasattr(v, 'shape'):
        print(f'  {k:12s} ({t}, shape={v.shape})')
    elif isinstance(v, (list, dict)):
        print(f'  {k:12s} ({t}, len={len(v)})')
    else:
        print(f'  {k:12s} ({t}) = {v}')

print('\n--- config ---')
pprint.pprint(r['config'])

print('\n--- ret (net) head 5 ---')
print(r['ret'].head())

print('\n--- gross_ret head 5 ---')
print(r['gross_ret'].head())

print('\n--- spy_ret head 5 ---')
print(r['spy_ret'].head())

print('\n--- weights head 5 (앞 6 컬럼만) ---')
print(r['weights'].iloc[:5, :6])

print('\n--- comp head 5 ---')
print(r['comp'].head())

print('\n--- meta head 5 ---')
print(r['meta'].head())

print(f'\n--- errors: {len(r["errors"])}건 ---')
if r['errors']:
    for e in r['errors'][:5]:
        print(' ', e)
```
